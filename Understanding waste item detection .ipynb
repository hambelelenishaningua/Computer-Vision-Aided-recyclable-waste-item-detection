```julia
### Data Loading
```

```julia

using Images, FileIO

function load_images(path)
    images = []
    
    for file in readdir(path)
        img = load(joinpath(path, file))
        img = imresize(img, (224,224))
        #  Converting RGB image to Float32 properly
        img = Float32.(channelview(img))  
        push!(images, img)
    end
    
    return images
end


data = load_images("C:\\Users\\hambe\\Downloads\\project\\detection items")
```

```julia
# Creating the patches
```

```julia
function create_patches(img, patch_size=10)
    patches = []

    for i in 1:patch_size:224
        for j in 1:patch_size:224
            patch = img[i:i+patch_size-1, j:j+patch_size-1, :]
            push!(patches, patch)
        end
    end

    return patches
end
```

```julia
## Context vs Target Masking
```

```julia
function mask_patches(patches)
    shuffled = shuffle(patches)
    split = Int(length(patches)*0.7)

    context = shuffled[1:split]
    target  = shuffled[split+1:end]

    return context, target
end
```

```julia
# Context Encoders
```

```julia
context_encoder = Chain(
    Conv((3,3), 3=>16, relu),
    MaxPool((2,2)),
    Conv((3,3), 16=>32, relu),
    Flux.flatten,
    Dense(5408,128)
)
```

```julia
# Target Encoder
```

```julia
target_encoder = deepcopy(context_encoder)
```

```julia
# Predictor
```

```julia
predictor = Chain(
    Dense(128,128,relu),
    Dense(128,128)
)
```

```julia
# Similarity Loss
```

```julia
function loss_fn(ctx, tgt)
    ctx_repr = predictor(context_encoder(ctx))
    tgt_repr = target_encoder(tgt)

    return Flux.Losses.mse(ctx_repr, tgt_repr)
end
```

```julia
# Training Loop
```

```julia
opt = ADAM(0.0003)

for epoch in 1:10
    for img in data

        patches = create_patches(img)
        context, target = mask_patches(patches)

        ctx = cat(context..., dims=4)
        tgt = cat(target..., dims=4)

        gs = gradient(() -> loss_fn(ctx, tgt), params(context_encoder, predictor))

        Flux.Optimise.update!(opt, params(context_encoder, predictor), gs)

    end

    println("Epoch $epoch complete")
end
```
